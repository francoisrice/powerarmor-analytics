```python
def fetchOptionContractPrices(symbol="",start_date=None):
    try: 
        # Fetch options contracts
        contractDF = obb.equity.price.historical(symbol, start_date=start_date, interval='1m', include_actions=False, provider="yfinance").to_df()
        contractPriceDF = calcContractAvgPrice(contractDF)
        return contractPriceDF
    except:
        print("Error fetching contract prices")

def calcContractAvgPrice(dataframe):
    dataframe['Price'] = (dataframe['high']+dataframe['low']+dataframe['close'])/3
    return dataframe['Price']
```
